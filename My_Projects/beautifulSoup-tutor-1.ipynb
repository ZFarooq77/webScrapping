```python
from bs4 import BeautifulSoup
import requests
```

```python
# Example usage with a URL
url = "https://www.immobiliare.it/annunci/108773485/"
```

```python
requests.get(url)
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
title = soup.find(class_='in-realEstateFeatures__title')
print(title)
```

```python
agency_div = soup.find('div', class_='nd-figure__image in-referent__image')
# Find the img tag and then get its 'alt' attribute
agency_name = agency_div.find('img', class_='nd-figure__content')['alt']
print(agency_name)

# print(title)
#div class="nd-figure__image in-referent__image"
```
